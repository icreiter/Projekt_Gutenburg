```python
# get bookID

url = 'https://www.gutenberg.org/browse/languages/zh'

res = req.get(url)

soup = bs(res.text, 'lxml')

ebook_links = soup.select(
    'div.pgdbbylanguage a[href^="/ebooks/"]'
)

ebook_ids = [link['href'].split('/')[2] for link in ebook_links]

print(ebook_ids)
```

```python
for id in ebook_ids:
    url = f'https://www.gutenberg.org/cache/epub/{id}/pg{id}.txt'

    res = req.get(url)

    if res.status_code == 200:
        content = res.text

        contCh = re.sub(r'[^\u4e00-\u9fa5\u3000-\u303F\uff00-\uffef]', '', content)

    file_path = os.path.join(folderPath, f'ebook_{id}.txt')
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(contCh)
    
    print(f'已存{id}.txt')
```
